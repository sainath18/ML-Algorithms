```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv("Logistic_Regression_Binary_Classsification_Churn_Data.csv")
data.head(2)
```

```python
data.info()
```

```python
data.keys()
```

```python
churn=pd.get_dummies(data["Churn"],drop_first=True)
churn.head(2)
```

```python
phone_service=pd.get_dummies(data["PhoneService"],drop_first=True)
phone_service.head(2)
```

```python
paperless_billing=pd.get_dummies(data["PaperlessBilling"],drop_first=True)
paperless_billing.head(2)
```

```python
cont=pd.get_dummies(data["Contract"],drop_first=True)
cont.head(2)
```

```python
method=pd.get_dummies(data["PaymentMethod"],drop_first=True)
method.head(2)
```

```python
df=pd.concat([data.MonthlyCharges,data.TotalCharges,data.tenure,paperless_billing,phone_service,cont,method,churn],axis=1)
df
```

```python
x=df.iloc[:,[0,2,3,4,5,6,7,8,9]].values
x
```

```python
x.shape
```

```python
y=df.iloc[:,-1].values
y
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=.20)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
# prediction
y_pred=lr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
```

```python

# confusion matrix
confusion_matrix(y_test,y_pred)
```

```python
# accuracy
accuracy_score(y_test,y_pred)
```

```python
# precision
precision_score(y_test,y_pred)
```

```python
# recall
recall_score(y_test,y_pred)
```

```python
# f1 score
f1_score(y_test,y_pred)
```
